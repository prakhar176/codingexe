```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-10,10,0.1)
print(x.shape)
```

```python
x2 = x ** 2 #square of x
```

```python
plt.plot(x,x2)
```

```python
x3 = x ** 3
```

```python
plt.figure(figsize=(15,7))
plt.plot(x, label='orignal', linestyle='--')
plt.plot(x2, label= 'squared', linestyle= ':')
plt.plot(x3, label='cube')
plt.legend() #activate the legends
plt.xlabel("data points") # add label to x axis
plt.ylabel("values of x")# add label to y axis
plt.title('The visualisation of x and its square and cube')
plt.show()
```

```python
cities = ['lucknow','jaipur','mumbai','pune','bangalore']
rainfall = [39,20,64,45,78]
```

```python
plt.bar(cities,rainfall)
```

```python
plt.barh(cities,rainfall)
```

```python
plt.figure(figsize=(11,8))
plt.bar(cities,rainfall,color ='grey',edgecolor = 'purple',hatch='//')
```

```python
plt.figure(figsize=(11,8))
plt.bar(cities,rainfall,color ='grey',edgecolor = 'purple',hatch='//', width =[.1,.2,.3,.4,.5])
```

```python
plt.figure(facecolor='white')
plt.pie(rainfall, labels=cities, autopct='%1.1f%%',radius=2)
plt.show()
```

```python
plt.figure(facecolor='white')
wedges = [0,0.3,0,0,0]
plt.pie(rainfall, labels=cities, autopct='%1.1f%%',radius=2,explode=wedges)
plt.show()
```

```python
prices = [213,34,46,678,45,67,5,78,34,6,23]
qty= [12,45,67,89,34,23,12,45,23,56,78]
quality =[12,45,68,73,99,33,85,67,77,72,44]
len(prices) == len(qty) ==len(quality)
```

```python
plt.scatter(prices,qty)
```

```python
plt.scatter(prices, qty,color='green',marker='x',s=100*3,alpha=.5)
plt.scatter(prices,quality,color='blue',marker='H',s=100*2,alpha=.5,edgecolors='black')
```

```python
x = np.random.randint(1,10,500)
y = np.random.randint(1,10,500)
plt.hist(x,bins=10,color='black',range=(0,10),align='left')
plt.hist(y,bins=10,color='blue',range=(0,10),align='left',alpha=.5)
plt.xticks(np.arange(0,11,1))
```

```python
plt.boxplot([prices,qty,quality])
plt.xticks(np.arange(1,4)[prices,qty,quality],rotation=45)
```
